```python
import numpy as np
import pandas as pd
from datetime import datetime
```

```python
def data_prep(fp):
    df = pd.read_csv(fp)
    
    df['post_expire'] = pd.to_datetime(df['post_expire'], utc=True)
    df['post_datetime'] = pd.to_datetime(df['post_datetime'], utc=True)
    
    df = df.loc[(df['post_datetime'] >= '2023-08-10')]

    return df

def plot(df):
    plot_df = df.groupby(df.post_datetime.dt.date).size()
    plot_df.plot.bar(xlabel='posting date', rot=25, fontsize=10)
    
```

```python
det_fp = './data/processed/craigslist_det_ev_sample.csv'
sf_fp = './data/processed/craigslist_sf_ev_sample.csv'
```

```python
plot(data_prep(sf_fp))
```

```python
plot(data_prep(det_fp))
```

```python
data_prep(det_fp).seller_type.value_counts().sort_index(ascending=False).plot.barh()
```

```python
data_prep(sf_fp).seller_type.value_counts().sort_index(ascending=False).plot.barh()
```
