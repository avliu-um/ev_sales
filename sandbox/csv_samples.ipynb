## CSV samples

```python
import json
import pandas as pd
```

```python
def json2csv(input_fp, output_fp):
    
    with open(input_fp, 'r') as read_file:
        data = json.load(read_file)
    
    data = pd.json_normalize(data)
    
    data.to_csv(output_fp, index=False)
```

```python
# config

kbb_json = '../kbb/data/data.json'
kbb_csv = '../sample_csv/kbb.csv'

craig_json = '../craigslist/data/data.json'
craig_csv = '../sample_csv/craig.csv'
```

```python
json2csv(kbb_json, kbb_csv)
json2csv(craig_json, craig_csv)
```
